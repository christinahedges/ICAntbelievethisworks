```python
from lightkurve import KeplerTargetPixelFile
```

```python
tpf = KeplerTargetPixelFile('../../NGC/data/Q1/2568982-targ.fits')
```

```python
tpf.plot(scale='log');
```

```python
tpf.to_lightcurve().plot();
```

```python
from sklearn.decomposition import FastICA, PCA
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = tpf.flux.reshape((tpf.flux.shape[0], (tpf.flux.shape[1]*tpf.flux.shape[2])))
```

```python
ncomp=1
ica = FastICA(n_components=ncomp)
S_ = ica.fit_transform(X)
A_ = ica.mixing_
```

```python
#for s in S_.T:
plt.plot(S_.T[0])
```

```python
plt.imshow(A_.reshape((10, 10, ncomp))[:,:,0], vmin=A_.min(), vmax=A_.max(), origin='bottom')
plt.colorbar()
```

```python
plt.imshow(A_.reshape((10, 10, ncomp))[:,:,1], vmin=A_.min(), vmax=A_.max())
plt.colorbar()
```

```python
plt.imshow(A_.reshape((10, 10, ncomp))[:,:,2], vmin=A_.min(), vmax=A_.max())
plt.colorbar()
```
